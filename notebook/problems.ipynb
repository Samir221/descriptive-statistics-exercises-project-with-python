# Descriptive statistics problems

### Exercise 1

We will use NumPy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
import numpy as np
import pandas as pd
from scipy.stats import skew
from scipy.stats import kurtosis

# Generate an array of 100 elements following a normal distribution
normal_array = np.random.normal(loc=0, scale=1, size=100)

# Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom
chi_square_array = np.random.chisquare(df=3, size=100)

# Calculate the main metrics and statistical measures
def calculate_metrics(array):
    metrics = {
        'mean': np.mean(array),
        'median': np.median(array),
        'variance': np.var(array),
        'standard deviation': np.std(array),
        'minimum': np.min(array),
        'maximum': np.max(array),
        '25th percentile': np.percentile(array, 25),
        '75th percentile': np.percentile(array, 75),
        'skewness': skew(array),
        'kurtosis': kurtosis(array),
    }
    return metrics

normal_metrics = calculate_metrics(normal_array)
chi_square_metrics = calculate_metrics(chi_square_array)

metrics_df = pd.DataFrame([normal_metrics, chi_square_metrics], index=['Normal Distribution', 'Chi-Square Distribution'])

print(metrics_df)
````

````python
import matplotlib.pyplot as plt

# Plot the histogram
plt.hist(normal_array, bins=20, edgecolor='black', alpha=0.7)
plt.title('Histogram of Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
````

````python
# Plot the histogram
plt.hist(chi_square_array, bins=20, edgecolor='black', alpha=0.7)
plt.title('Histogram of Chi-Square Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
````

````python
print(f"NORMAL ARRAY: {normal_array}")
print(f"CHI SQUARE ARRAY: {chi_square_array}")
````

### Exercise 2

Write a Python program to calculate the standard deviation of the following data:

```py
data = [4, 2, 5, 8, 6]
```

````python
# TODO

data = [4, 2, 5, 8, 6]

data_mean = np.mean(data)

deviation_sum = 0
for item in data:
    deviation_sum = deviation_sum + (item - data_mean)**2 

div_deviation = deviation_sum / (len(data) -1)
standard_deviation = np.sqrt(div_deviation)

print(f"Standard Deviation is: {standard_deviation}")
````
